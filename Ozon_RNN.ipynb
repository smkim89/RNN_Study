```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from keras.layers import *
from keras.models import *
from keras.utils import *
from sklearn.preprocessing import *
```

```python
label = ['feat{}'.format(i) for i in range(73)]

df = pd.read_csv('eighthr.data', names= label)
df
```

```python
#피쳐의 데이터 타입을 볼수 있다.
df.dtypes

#df에 object들을 숫자로 바꿔주는 함수. coerce는 숫자로 바꿀수 없을때 NaN으로 처리한다.
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
#Nan을 버린다.
df.dropna(inplace=True)
df
```

```python
Y = df['feat72']
Y
```

```python
Y = to_categorical(Y)
Y.shape
```

```python
X = df.drop(['feat72'], axis=1)

X_train = np.asarray(X[:-100].values.tolist(), dtype=np.float64)
X_test = np.asarray(X[-100:].values.tolist(), dtype=np.float64)
Y_train = Y[:-100]
Y_test =Y[-100:]
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
X_train = X_train[:1700]
Y_train = Y_train[:1700]
```

```python
#(sample, timestep, feature)
#10으로 나누기위해 위에를 1700까지만 처리한다.
X_train = X_train.reshape(-1,10,72)
Y_train = Y_train.reshape(-1,10,2)
```

```python
#데이터가 170개로 변경한다 (timestep을 10개를넣었기때문에 1700/10)
X_train.shape
```

```python
Y_train.shape
```

```python
model = Sequential()
#input_shape(timestep, feature)
#LSTM에 층층히 쌓으려면 return_sequences=True로 해야한다.
model.add(LSTM(128, input_shape=(10, 72), return_sequences= True))
model.add(LSTM(256, return_sequences= True))
model.add(Dense(2, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
model.fit(X_train, Y_train, epochs=10, batch_size=1 , validation_split=0.1)
```

```python
Y_test = Y_test.reshape(-1, 10, 2)
X_test = X_test.reshape(-1,10,72)
```

```python
score = model.evaluate(X_test, Y_test)
print(score)
```
